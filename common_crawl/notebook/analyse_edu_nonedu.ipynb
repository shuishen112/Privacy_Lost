```python
import pandas as pd 

df = pd.read_csv("../resource/url_index.txt",sep = "\t")
df.head()
```

```python
df['rank_gap'] = df['control_url_index'] - df['edu_url_index']
df['rank_gap'].describe([0.99])
```

```python
df['rank_gap'].value_counts()
```

```python
import matplotlib.pyplot as plt
import seaborn  as sns
plt.figure(figsize=(13,10), dpi= 200)
plt.rc('xtick', labelsize=20)    # fontsize of the tick labels
plt.rc('ytick', labelsize=20)    # fontsize of the tick labels
g2 = sns.boxplot(df['rank_gap'], notch=False)
g2.set(xlabel = None)
g2.set(ylabel = None)
```
