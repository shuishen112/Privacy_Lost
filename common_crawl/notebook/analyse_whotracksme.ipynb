```python
import os 
name_list = sorted(os.listdir("../resource/whotracksme"))

trackers_num = []
names = []
for name in name_list:
    if len(name) == 7:
        names.append(name)
        df_trackers = pd.read_csv(f"../resource/whotracksme/{name}/global/trackers.csv")
        trackers_num.append(len(df_trackers['tracker'].unique()))
```

```python
import matplotlib.pyplot as plt
plt.plot(names, trackers_num,marker='o',label = name_list, linestyle=':',linewidth = 2)
plt.rcParams["figure.figsize"] = (20,10)
plt.xticks(rotation=90)
plt.show()
```
