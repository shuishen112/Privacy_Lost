```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np 
SMALL_SIZE = 12
MEDIUM_SIZE = 10
BIGGER_SIZE = 20

plt.rc('font', size=SMALL_SIZE)          # controls default text sizes
plt.rc('axes', titlesize=SMALL_SIZE)     # fontsize of the axes title
plt.rc('axes', labelsize=BIGGER_SIZE)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=BIGGER_SIZE)    # fontsize of the tick labels
plt.rc('ytick', labelsize=BIGGER_SIZE)    # fontsize of the tick labels
plt.rc('legend', fontsize=SMALL_SIZE)    # legend fontsize
plt.rc('figure', titlesize=BIGGER_SIZE)  # fontsize of the figure title
```

# Categorical evaluation


```python
df_merge = pd.read_csv("../dataset_archive/df_category_rate_compare_whoexclude_2012.csv")
for year in range(2013, 2022):
    df = pd.read_csv(f"../dataset_archive/df_category_rate_compare_whoexclude_{year}.csv")
    df_merge = df_merge.merge(df, how="left", on = "category",suffixes = ["",year])
df_merge = df_merge.fillna(0)
df_merge
```

```python
plt.plot(df_merge['category'],df_merge["edu"],marker='o',label = "2012", linestyle=':',linewidth = 2)
# plt.plot(df_merge['category'],df_merge["edu2013"],marker='o',label = "2013", linestyle=':',linewidth = 2)
for year in range(2013,2022):
    plt.plot(df_merge['category'],df_merge[f"edu{year}"],marker='o',label = year, linestyle=':',linewidth = 2)
# plt.plot(df_merge['category'],df_merge["no-edu"],marker='o',label = "edu", linestyle=':',linewidth = 2)
plt.xticks(rotation=90)
plt.legend()
```

```python
plt.plot(df_merge['category'],df_merge["no-edu"],marker='o',label = "2012", linestyle=':',linewidth = 2)
# plt.plot(df_merge['category'],df_merge["edu2013"],marker='o',label = "2013", linestyle=':',linewidth = 2)
for year in range(2013,2022):
    plt.plot(df_merge['category'],df_merge[f"no-edu{year}"],marker='o',label = year, linestyle=':',linewidth = 2)
# plt.plot(df_merge['category'],df_merge["no-edu"],marker='o',label = "edu", linestyle=':',linewidth = 2)
plt.xticks(rotation=90)
plt.legend()
```

# draw the category over the year

# draw the figure for different categories

```python
import matplotlib.lines as mlines
import numpy as np 
# Import Data
df_edu = df_edu[:8]
df_edu = df_edu.replace({'category': {"Customer Interaction": "Customer", "Audio Video Player": "Audio"}})
left_label = [str(c) + ', '+ str(round(y)) for c, y in zip(df_edu.category, df_edu['edu2012'])]
right_label = [str(c) + ', '+ str(round(y)) for c, y in zip(df_edu.category, df_edu['edu2021'])]
klass = ['red' if (y1-y2) < 0 else 'green' for y1, y2 in zip(df_edu['edu2012'], df_edu['edu2021'])]

# draw line
# https://stackoverflow.com/questions/36470343/how-to-draw-a-line-with-matplotlib/36479941
def newline(p1, p2, color='black'):
    ax = plt.gca()
    l = mlines.Line2D([p1[0],p2[0]], [p1[1],p2[1]], color='red' if p1[1]-p2[1] > 0 else 'green', marker='o', markersize=6)
    ax.add_line(l)
    return l

fig, ax = plt.subplots(1,1,figsize=(14,12), dpi= 80)

# Vertical Lines
ax.vlines(x=1, ymin=0, ymax=1, color='black', alpha=0.7, linewidth=2, linestyles='dotted')
ax.vlines(x=3, ymin=0, ymax=1, color='black', alpha=0.7, linewidth=2, linestyles='dotted')

# Points
ax.scatter(y=df_edu['edu2012'], x=np.repeat(1, df_edu.shape[0]), s=10, color='black', alpha=0.7)
ax.scatter(y=df_edu['edu2021'], x=np.repeat(3, df_edu.shape[0]), s=10, color='black', alpha=0.7)

# Line Segmentsand Annotation
for p1, p2, c in zip(df_edu['edu2012'], df_edu['edu2021'], df_edu['category']):
    newline([1,p1], [3,p2])
    ax.text(1-0.05, p1, c + ', ' + str(round(p1,3)), horizontalalignment='right', verticalalignment='center', fontdict={'size':15})
    ax.text(3+0.05, p2, c + ', ' + str(round(p2,3)), horizontalalignment='left', verticalalignment='center', fontdict={'size':15})

# 'Before' and 'After' Annotations
ax.text(1-0.05, 0.6, 'BEFORE', horizontalalignment='right', verticalalignment='center', fontdict={'size':18, 'weight':700})
ax.text(3+0.05, 0.6, 'AFTER', horizontalalignment='left', verticalalignment='center', fontdict={'size':18, 'weight':700})

# Decoration
ax.set_title("The Usage Rate Change of Different Categories Trackers ", fontdict={'size':22})
ax.set(xlim=(0,4), ylim=(0,0.7))
ax.set_xticks([1,3])
ax.set_xticklabels(["2012", "2021"])
plt.yticks(np.arange(0, 0.7, 0.1), fontsize=20)

# Lighten borders
plt.gca().spines["top"].set_alpha(.0)
plt.gca().spines["bottom"].set_alpha(.0)
plt.gca().spines["right"].set_alpha(.0)
plt.gca().spines["left"].set_alpha(.0)
plt.show()
```

```python
# get the categories of trackers
import tldextract
df_trackers_categories = pd.read_csv("../resource/new_trackerList_with_categories.txt",sep = "\t",names = ['tracker_domain','tracker_name','category','company'])
df_trackers_categories.head()

def get_domain(row):
    return tldextract.extract(row['tracker_domain']).domain

df_trackers_categories['tracker_domain'] = df_trackers_categories.apply(get_domain, axis = 1)

df_trackers_categories = df_trackers_categories.drop_duplicates()
df_trackers_categories.head()
```

```python
tld_categories = dict(zip(df_trackers_categories["tracker_domain"].to_list(), df_trackers_categories["category"].to_list()))
tld_categories
```

```python
df_edu_merge = df_edu.merge(df_trackers_categories, how = "left", on = "tracker_domain")
df_edu_merge[df_edu_merge.category == 'Advertising'].head()
```

```python
# plot the change of Advertising trackers
years = list(range(2012, 2022))
df_edu_merge['category'].fillna("Unkonwn").unique()

for category in df_edu_merge['category'].fillna("Unkonwn").unique():
    
    tracker_list = df_edu_merge[df_edu_merge.category == category]['tracker_domain'][:10]
    for trackers in tracker_list:
        rate_edu_l = df_edu_merge.loc[df_edu_merge.tracker_domain == trackers].values.flatten().tolist()

        plt.plot(years,rate_edu_l[1:-3],marker='o',label = trackers, linestyle=':',linewidth = 2)
    plt.xlabel('Year', fontsize=14)
    plt.ylabel('tracker rate in edu websites', fontsize=14)
    # plt.yticks(np.arange(0, 0.7, step=0.1))
    plt.title("edu website " + category)
    plt.legend()
    plt.show()
    # plt.savefig("../images/section4/tracker_rate_change_edu.png",dpi = 200)
```
