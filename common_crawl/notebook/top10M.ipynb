```python
import pandas as pd 

df = pd.read_csv("../resource/top10million10000_timegap.csv")
df.head()
```

```python
df = df.dropna()
len(df)
```

```python
def get_age(row):
    year_gap = int(str(row['new_edu'])[:4]) - int(str(row['old_edu'])[:4])
    return year_gap

df["year_gap"] = df.apply(get_age, axis = 1)
```
