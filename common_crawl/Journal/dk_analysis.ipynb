```python
import pandas as pd 
```

```python
df_top_10m = pd.read_csv("top_10m_2017.txt",sep = "\t")
df_top_10m.head()
```

```python
# df_top_10m = df_top_10m.sort_values(by='#pr_pos')
# df_top_10m
def reversed_domain(row):
    return ".".join(row["#host_rev"].split(".")[::-1])
df_top_10m["reversed_domain"] = df_top_10m.apply(reversed_domain, axis = 1)
df_top_10m
```

```python
def get_tld(row):
    return row['reversed_domain'].split(".")[-1]

df_top_10m["tld"] = df_top_10m.apply(get_tld, axis = 1)
df_top_10m.head()
```

```python
df_top_dk = df_top_10m[df_top_10m.tld == "dk"]
```

```python
import os
if not os.path.exists("dk/"):
    os.mkdir("dk")
df_top_dk.to_csv("dk/dk_2017.csv", index = None, sep = "\t")
```
